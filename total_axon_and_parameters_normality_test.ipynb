```python
#Total axon both nerves - Normality test
```

```python
data = pd.read_csv('/home/onarvaez/Documents/test_normalidad_isq/total_axon_intact_degenerated.csv')

vector_intact_degenerated=[]
for i in range(100):
    vector_intact_degenerated.append(data.iloc[i,0])
    vector_intact_degenerated.append(data.iloc[i,1])

result_intact_degenerated=sst.normaltest(vector_intact_degenerated)
```
